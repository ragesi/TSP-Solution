```python
import numpy as np
import math

# set up the initial space
# n = log2(N), where N is dataset size
# t = n + round up(2 + log2(1/2*e0) where e0 is the error rate of phase estimation, 
# we will choose e0 = 0.001

zero_state = np.array([1,0])
one_state = np.array([0,1])
single_dim_identity = np.eye(2)
# 外积
zero_single_dim_projector = np.outer(zero_state,np.transpose(zero_state))
one_single_dim_projector = np.outer(one_state,np.transpose(one_state))
pauli_X = np.array([[0,1],[1,0]])
hadamard = np.array([[1,1],[1,-1]])*1/math.sqrt(2)

identity_power = lambda k: kronecker_power(single_dim_identity,k)

def unitary_initilization(qc,t,n):
    """
    This puts both lower 'n' registers entangled with each other
    一共有三个量子寄存器，第一个有t个量子比特，第二个和第三个有n个量子比特
    和那个U_in算符一样，第二个和第三个的量子态一致。
    """
    # entangle the states
    for i in range(n):
        qc.h(t+i)
        qc.cx(t+i,t+n+i)
    return qc

def kronecker_power(matrix, k):
    """ matrix自身进行k次张量积 """
    import numpy as np
    new_matrix = matrix.copy()
    for _ in range(k-1):
        # 张量积
        new_matrix = np.kron(new_matrix,matrix)
    return new_matrix

def get_specific_operator_for_iteration(counting_qubit, unitary_l_norm_op, t, n):
    """
    构造QPE算法中的control-U算符
    Apply specific operator for given iteration within quantum phase estimation algorithm.
    """
    zero_control = None
    one_control = None

    if counting_qubit == 0:
        # (zero_single_dim_projector) * (Identity ^* (t+n-counting_qubit-1))
        zero_control = np.kron(zero_single_dim_projector,identity_power(t+n-counting_qubit-1))
        # (one_single_dim_projector) * (Identity ^* (t-counting_qubit-1)) * Unitary_to_apply
        one_control = np.kron(np.kron(one_single_dim_projector,identity_power(t-counting_qubit-1)), unitary_l_norm_op)
    else:
        common_iden = identity_power(counting_qubit)
        # (Identity ^* counting_qubit) * (zero_single_dim_projector) * (Identity ^* (t+n-counting_qubit-1))
        zero_control = np.kron(np.kron(common_iden,zero_single_dim_projector),identity_power(t+n-counting_qubit-1))
        if counting_qubit == t-1:
            # (Identity ^* counting_qubit) * (one_single_dim_projector) * Unitary_to_apply
            one_control = np.kron(np.kron(common_iden,one_single_dim_projector), unitary_l_norm_op)
        else:
            # (Identity ^* counting_qubit) * (one_single_dim_projector) * (Identity ^* (t-counting_qubit-1)) * Unitary_to_apply
            one_control = np.kron(np.kron(np.kron(common_iden,one_single_dim_projector),identity_power(t-counting_qubit-1)), unitary_l_norm_op)
    
    assert zero_control.shape == one_control.shape
    
    return zero_control + one_control

def qsc_multi_qubit_unitary_phase_estimation(qc,l_norm,t,n):
    from matrix_exp import get_exponential_unitary
    from qiskit.circuit.library import QFT
    """
    得到拉普拉斯算子并应用在QPE算法中
    Recieves normalized laplacian and outputs phase estimation algorithm
    """
    for i in range(t):
        qc.h(i)
    
    #Build the giant control unitary
    unitary_l_norm_op = get_exponential_unitary(l_norm)    


    active_qubits = [i for i in range(t+n)]

    repetitions = 1
    for counting_qubit in range(t):
        
        # get the corresponding CNOT operation for the given iteration
        unitary_control_op = get_specific_operator_for_iteration(counting_qubit,unitary_l_norm_op,t,n)

        for i in range(repetitions):
            # apply the control-U
            qc.unitary(unitary_control_op, qubits=active_qubits)
        repetitions *= 2
    
    estimation_qubits = [i for i in range(t)]
    qc = qc.compose(QFT(num_qubits=t, inverse=True), estimation_qubits)
    # # Measure of course!
    # for c in estimation_qubits:
    #     qc.measure(c,c)
    
    return qc


def bottom_up_quantum_phase_estimation(qc, estimation_qubits=[], qubit_to_estimate=0):
    from qiskit.circuit.library import QFT
    import math
    if len(estimation_qubits)==0:
        estimation_qubits = [i+1 for i in range(qc.num_qubits-1)] # just the last couple
    # Apply H-Gates to counting qubits:
    for qubit in estimation_qubits:
        qc.h(qubit)
    print(f"inside function qc.num_clbits:{qc.num_clbits}")
    # Prepare our eigenstate |psi>:
    qc.x(qubit_to_estimate)

    # Do the controlled-U operations:
    angle = 2*math.pi/3
    repetitions = 1
    for counting_qubit in estimation_qubits[::-1]: # reverse
        for _ in range(repetitions):
            qc.cp(angle, counting_qubit, qubit_to_estimate);
        repetitions *= 2

    # Do the inverse QFT:
    qc = qc.compose(QFT(num_qubits=len(estimation_qubits), inverse=True), estimation_qubits)
    # Measure of course!
    for c in estimation_qubits:
        qc.measure(c,c)

    return qc

def unitary_qft(n):
    # acting on n number of states
    import math
    summation = np.zeros((2**n,2**n)) # zero matrix

    for j in range(2**n):
        for k in range(2**n):
            theta = 2*math.pi*j*k/(2**n)
            w_jk_N = math.cos(theta) + 1j*math.sin(theta) # e^i*theta = cos(theta) + i*sin(theta)
            k_vector = binary_decimal_vector_representation(get_binary_decimal_representation(k,n))
            j_vector = binary_decimal_vector_representation(get_binary_decimal_representation(j,n))
            summation += w_jk_N * np.outer(k_vector,j_vector)
    
    return summation / (math.sqrt(2**n))

def get_unitary_from_phase_estimation(t,n, l_norm):
    # this unitary acts on t+n space
    U_pe = get_conjugate_transpose(unitary_qft(t))
    
    #Build the giant control unitary
    from matrix_exp import get_exponential_unitary
    unitary_l_norm_op = get_exponential_unitary(l_norm)

    repetitions = 2**t
    for counting_qubit in range(t-1,-1,-1):
        
        # get the corresponding CNOT operation for the given iteration
        unitary_control_op = get_specific_operator_for_iteration(counting_qubit,unitary_l_norm_op,t,n)

        for i in range(repetitions):
            # apply the control-U
            U_pe *= unitary_control_op
        repetitions //= 2
    
    hadamard_tensors = np.kron(kronecker_power(hadamard,t), identity_power(n))

    U_pe *= hadamard_tensors

    return U_pe

def NormalizedLaplaceTransform(W):
    # print(f"W orginial: {W}")
    diag_elements = np.diag(np.diag(W))
    W = W-diag_elements
    # print(f"W with diagonals as zero:{W}")
    diagment = np.sum(W,axis=1)
    D=np.diag(diagment)
    # print(f"D:{D}")
    L = D-W
    # print(f"L:{L}")
    D_sqrt = np.sqrt(np.linalg.inv(D))
    # print(f"D_sqrt:{D_sqrt}")
    l_norm = np.matmul(D_sqrt,np.matmul(L,D_sqrt))
    # print(f"l_norm:{l_norm}")
    return l_norm

def get_conjugate_transpose(np_matrix):
    transpose_mat = np.matrix.transpose(np_matrix)
    return np.conjugate(transpose_mat)

def get_adjacency_and_weight_matrix(raw_data, n_nearest_neighbors):
    n = len(raw_data)
    
    from sklearn.neighbors import NearestNeighbors
    nbrs = NearestNeighbors(n_neighbors=n_nearest_neighbors, algorithm='auto').fit(raw_data)
    distances, indices = nbrs.kneighbors(raw_data)

    from sklearn.metrics.pairwise import rbf_kernel
    rbf = rbf_kernel(raw_data) # input must be np.array

    # only add weights if indices correspond to a connection
    weight_matrix = np.zeros((n,n))
    for i in range(len(indices)):
        for j in indices[i]:
            weight_matrix[i,j] = rbf[i][j]
    
    adjacency = nbrs.kneighbors_graph(raw_data).toarray()
    return adjacency, weight_matrix

def get_binary_decimal_representation(x, n):
    # get reverse binary of length 'n'
    # ex: x = 1, n = 3 => returns '001'
    binary_format = f"{{0:0{n}b}}"
    return binary_format.format(x)

def get_integer_from_binary(b):
    return int(b, 2)

def get_binary_approximation_of_lambda_0(lambda_0, t):
    # finding binary estimation of lambda_0 which first guess is at = 1/(2*N)
    best_aprox = 0
    best_binary_aprox = get_binary_decimal_representation(0, t)
    for i in range(2**t):
        binary_decimal_str = get_binary_decimal_representation(i, t)
        diviser = 2
        decimal_num = 0
        for i,char in enumerate(binary_decimal_str):
            if char == '1':
                decimal_num += 1/diviser
            diviser *= 2
        # assign the new best approximation since i increases over time and thus decimal increases too over iterations
        best_aprox = decimal_num
        best_binary_aprox = binary_decimal_str
        if decimal_num >= lambda_0:
            break
    return best_aprox, best_binary_aprox

def binary_decimal_vector_representation(binary_str):
    assert len(binary_str) > 1
    import numpy as np
    state = zero_state.copy() if binary_str[0] == '0' else one_state.copy()
    # top one will be 1/2 and bottm one will be 1/2^t
    for c in binary_str[1:]:
        tensor_state = zero_state.copy() if c == '0' else one_state.copy()
        state = np.kron(state, tensor_state)
    return state

def binary_decimal_outer_product(binary_str1, binary_str2=''):
    binary_str2 = binary_str1.copy() if binary_str2 == '' else binary_str2
    # constructs | binary_str1 >< binary_str2 | in matrix form
    assert len(binary_str1) > 1 and len(binary_str2) > 1 and len(binary_str1) == len(binary_str2)
    state1 = binary_decimal_vector_representation(binary_str1)
    state2 = binary_decimal_vector_representation(binary_str2)
    return np.outer(state1,state2)

def create_O_f_oracle_in_unitary_form(N, t):
    import numpy as np
    lambda_0 = 1/(2*N)
    decimal_aprox, binary_aprox = get_binary_approximation_of_lambda_0(lambda_0, t)
    final_unitary = np.zeros((N,N))
    # finding binary estimation of lambda_0 which first guess is at = 1/(2*N)
    for i in range(2**t):
        binary_decimal_str = get_binary_decimal_representation(i, t)
        diviser = 2
        decimal_num = 0
        for i,char in enumerate(binary_decimal_str):
            if char == '1':
                decimal_num += 1/diviser
            diviser *= 2
        if decimal_num >= decimal_aprox:
            final_unitary += binary_decimal_outer_product(binary_decimal_str)
        else: # we subtract the ones that are less than the decimal approximation
            final_unitary -= binary_decimal_outer_product(binary_decimal_str)
    return final_unitary

def create_phase_unitary_from_grover(t,n):
    # any state |0> -> |0>
    # all other states |x> -> |x> for x>0
    t_zeros_plus_one = np.zeros((2**t,2**t))
    t_zeros_plus_one[0][0] = 1
    n_zeros_plus_one = np.zeros((2**n,2**n))
    n_zeros_plus_one[0][0] = 1
    t_phase = 2*t_zeros_plus_one - np.eye(2**t)
    n_phase = 2*n_zeros_plus_one - np.eye(2**n)

    return np.kron(np.kron(t_phase,n_phase),n_phase)

def get_inverse_of_unitary_initilization(n):
    # this acts on n x n qubits
    import math
    import numpy as np
    final_unitary = np.kron(kronecker_power(hadamard, n),identity_power(n))
    common_CNOT_operator = np.kron(zero_single_dim_projector, identity_power(n)) + np.kron(np.kron(one_single_dim_projector, identity_power(n-1)),pauli_X)
    for i in range(n):
        if i == 0:
            final_unitary *= np.kron(common_CNOT_operator, identity_power(n-i-1))
        elif i == n-1:
            final_unitary *= np.kron(identity_power(i),common_CNOT_operator)
        else:
            final_unitary *= np.kron(identity_power(i),np.kron(common_CNOT_operator, identity_power(n-i-1)))
    return final_unitary

def create_U_inv_from_govers_algorithm(t,n, l_norm):
    U_inv = np.kron(get_unitary_from_phase_estimation(t,n,l_norm),identity_power(n))

    U_inv *= np.kron(identity_power(t),get_conjugate_transpose(get_inverse_of_unitary_initilization(n)))

    U_inv *= create_phase_unitary_from_grover(t,n)

    U_inv *= np.kron(identity_power(t),get_inverse_of_unitary_initilization(n))

    U_inv *= np.kron(get_conjugate_transpose(get_unitary_from_phase_estimation(t,n,l_norm)),identity_power(n))

    return U_inv

def grovers_unitary(N,t,n,l_norm):
    O_f = np.kron(create_O_f_oracle_in_unitary_form(N, t),identity_power(n+n))

    return create_U_inv_from_govers_algorithm(t,n,l_norm) * O_f
```

```python
N = [(i+1)*1000000000000000000 for i in range(10000)]
k = 3
e0 = 0.01
c_complexity = [k*(j**3) for j in N]



import matplotlib.pyplot as plt

plt.plot(N,c_complexity,label='classical')

for i in range(16,19,1):
    q_complexity = [math.sqrt(k*(j**3))*(math.log2(j)**i) for j in N]
    plt.plot(N,q_complexity,label=f'quantum where poly(logn) = log^{i}')
plt.xlabel('N data points')
plt.ylabel('time complexity')
plt.title(f'N data points vs time complexity w/ taking d^4/e0 = 1 (constant)')
plt.legend()
plt.show()
```
